```python
from bs4 import BeautifulSoup
import pandas as pd, re

brand = "더샘"
product_name = "샘물 싱글 섀도우"

# 불필요한 텍스트 제거 함수
def clean_text(t):
    t = re.sub(r"\[.*?\]", "", t)
    t = re.sub(r"\(.*?\)", "", t)
    t = re.sub(r"품절", "", t)
    t = re.sub(r"\s+", " ", t)
    return t.strip(" -|/").strip()

rows = []
# 핵심 수정: li[data-key]로 직접 탐색
for li in soup.select("li[data-key]"):
    raw = li["data-key"]
    shade = clean_text(raw)

    # 가격 추출
    price_tag = li.select_one("span.tx_num")
    price = price_tag.get_text(strip=True) if price_tag else ""

    # 이미지 URL 추출
    img = li.select_one("img")
    swatch = img.get("src") if img else ""
    if swatch.startswith("//"):
        swatch = "https:" + swatch

    rows.append({
        "brand": brand,
        "product_name": product_name,
        "category": "eye",
        "shade_name": shade,
        "price": price,
        "finish": "",
        "image_url": "",
        "swatch_url": swatch
    })

# CSV 저장
filename = f"{brand}_{product_name}".replace(" ", "_") + ".csv"
df = pd.DataFrame(rows).drop_duplicates(subset=["shade_name"]).reset_index(drop=True)
df.to_csv(filename, index=False, encoding="utf-8-sig")

print(f"✅ CSV 생성 완료: {filename}")
print(df.head(10))
```
